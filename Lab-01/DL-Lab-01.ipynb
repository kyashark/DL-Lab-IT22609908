**2) Compare Exponential, Uniform, Normal Distributions**

```python
# Create arrays
exp_data = np.random.exponential(scale=1.0, size=(100000, 1))
uni_data = np.random.rand(100000, 1)       # Uniform
norm_data = np.random.normal(size=(100000, 1))  # Normal
```

```python
# Plot histograms
plt.figure(figsize=(8,5))
plt.hist(exp_data, density=True, bins=100, histtype="step", label="Exponential")
plt.hist(uni_data, density=True, bins=100, histtype="step", label="Uniform")
plt.hist(norm_data, density=True, bins=100, histtype="step", label="Normal")

# Customize
plt.xlim(-2, 6)
plt.ylim(0, 1.5)
plt.legend(loc="upper right")
plt.title("Random Distributions Comparison")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

**3) 3D Plot of Z = XÂ² + YÂ²**

```python
# Create X and Y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z
Z = X**2 + Y**2
```

```python
# Make 3D plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

**4) Pearson & Spearman Correlation Heatmaps**

```python
# Load dataset from seaborn tutorial
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv

# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
print(df.head())
```

```python
# Select only the required columns
stats = df[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
print(stats.head())
```

```python
# Calculate correlations
pearson_corr = stats.corr(method="pearson")
spearman_corr = stats.corr(method="spearman")

print("Pearson Correlation:\n", pearson_corr)
print("\nSpearman Correlation:\n", spearman_corr)
```

```python
# Visualize with heatmaps
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Correlation")

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('Au_nanoparticle_dataset (1).csv')
df.head()
```

**1) Filter columns**

```python
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

**2) Display the first 20 samples**

```python
new_df.head(20)
```

**3) Calculate statistics**

```python
# Mean
mean_values = new_df.mean()

# Standard deviation
std_values = new_df.std()

# Quartiles
quartiles = new_df.quantile([0.25, 0.5, 0.75])

print("Mean values:\n", mean_values)
print("\nStandard deviation:\n", std_values)
print("\nQuartiles:\n", quartiles)
```

**4) Plot histograms in 1x4 layout**

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

for i, col in enumerate(new_df.columns):
    axes[i].hist(new_df[col], bins=20, color='skyblue', edgecolor='black')
    axes[i].set_title(col)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**5) Pairplot using Seaborn**

```python
sns.pairplot(new_df)
plt.show()
```

**6) PairGrid with custom diagonal and lower plots**

```python
# new_df contains only the 4 features
g = sns.PairGrid(new_df)
```

```python
# Upper triangle: histograms
g.map_upper(sns.histplot)
```

```python
# Diagonal: histogram + KDE
g.map_diag(lambda x, **kwargs: sns.histplot(x, kde=True, **kwargs))
```

```python
# Lower triangle: bivariate KDE
g.map_lower(sns.kdeplot, fill=True)
```
